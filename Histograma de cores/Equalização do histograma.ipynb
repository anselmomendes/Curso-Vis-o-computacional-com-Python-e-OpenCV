### importando bibliotecas

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#argunemto imread: 0 colorida, 1 / 1 com alfpha / -1 preto e branca;
img = cv.imread("einstein.jpg", 0)
```

```python
imgEqualizada = cv.equalizeHist(img)
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray)

ax1 = fig.add_subplot(122)
plt.imshow(imgEqualizada, cmap=plt.cm.gray)
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.hist(img.ravel(), 256, [0, 256])

ax1 = fig.add_subplot(122)
plt.hist(imgEqualizada.ravel(), 256, [0, 256])
plt.show()
```

##### equalizando imagem colorida

```python
img2 = cv.imread("einstein.jpg", 1)
```

```python
R = img2[:,:, 0]
G = img2[:,:, 1]
B = img2[:,:, 2]
```

```python
imgEqualizadaR = cv.equalizeHist(R)
imgEqualizadaG = cv.equalizeHist(G)
imgEqualizadaB = cv.equalizeHist(B)
```

```python
img3 = cv.merge((imgEqualizadaR, imgEqualizadaG, imgEqualizadaB))
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.imshow(img2)

ax1 = fig.add_subplot(122)
plt.imshow(img3)
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
plt.hist(img2.ravel(), 256, [0, 256])

ax1 = fig.add_subplot(122)
plt.hist(img3.ravel(), 256, [0, 256])
plt.show()
```
